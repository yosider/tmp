```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f(x) as given in the problem
def f(x):
    return (
        4 * (1 - np.sign(np.abs(x - 22)))
        + 16 * (1 - np.sign(np.abs(x - 37)))
        + 17 * (1 - np.sign(np.abs(x - 2)))
        + 21
        * (
            7
            - np.sign(np.abs(x - 12))
            - np.sign(np.abs(x - 16))
            - np.sign(np.abs(x - 17))
            - np.sign(np.abs(x - 27))
            - np.sign(np.abs(x - 32))
            - np.sign(np.abs(x - 42))
            - np.sign(np.abs(x - 47))
        )
        + 23
        * (
            3
            - np.sign(np.abs(x - 11))
            - np.sign(np.abs(x - 28))
            - np.sign(np.abs(x - 33))
        )
        + 28 * (2 - np.sign(np.abs(x - 21)) - np.sign(np.abs(x - 36)))
        + 29
        * (
            3
            - np.sign(np.abs(x - 13))
            - np.sign(np.abs(x - 26))
            - np.sign(np.abs(x - 46))
        )
        + 31
        * (
            10
            - np.sign(np.abs(x - 1))
            - np.sign(np.abs(x - 3))
            - np.sign(np.abs(x - 7))
            - np.sign(np.abs(x - 18))
            - np.sign(np.abs(x - 23))
            - np.sign(np.abs(x - 31))
            - np.sign(np.abs(x - 38))
            - np.sign(np.abs(x - 41))
            - np.sign(np.abs(x - 43))
            - np.sign(np.abs(x - 48))
        )
    )
```

```python
# Set up the grid for plotting
x_values = np.arange(0, 50 + 0.1, 0.1)  # x goes from 0 to 50
y_values = np.arange(0, 5 + 0.1, 0.1)  # y goes from 0 to 5
X, Y = np.meshgrid(x_values, y_values)
a = np.pi
```

```python
def z(x, y, a):
    return (
        np.maximum(
            np.floor(
                np.mod(
                    f(
                        5 * np.floor(np.mod(a * 10 ** np.floor(x / 5), 10))
                        + np.mod(np.floor(x), 5)
                    )
                    * 2 ** (-np.floor(y)),
                    2,
                )
            ),
            1 - np.abs(np.floor(x) - 5) - np.abs(np.floor(y)),
        )
        > 0.5 + 0**x
    )
```

```python
Z = z(X, Y, a)
```

```python
# Plotting
plt.figure(figsize=(10, 2))
plt.contourf(X, Y, Z, levels=1, cmap="Greys")
plt.title("Plot of the complex formula on the x-y plane")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()
plt.show()
```
